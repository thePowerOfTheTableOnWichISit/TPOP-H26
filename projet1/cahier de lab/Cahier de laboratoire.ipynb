Notre première idée est de trouver le temps que prend le laser pour être détecté à la sortie du laser, puis de trouver le temps à une distance puis de soustraire le temps trouvé. La mesure à 7.5 cm est de 36800 ns, la grande différence provient srm de trucs weird, on va reprendre une mesure au loins

On voulais utilisé le sync out du générateur de fonction pour savoir quand est-ce le signal est haut, cependant il y a une différence d'environ 10 ns entre la montée du sync et le signal en sortie. On à donc décider de mesuré la monté du signal directement à l'aide d'un connecteur en T.

Le temps entre la montée du signal envoyé et la détection du laser est d'environ 1.5 ms. Ceci porte à croire qu'un temps non négligeable entre avant que le laser envoie son load. deux hypothèse sont émise. Premièrement, la tension monte sur une plage de 6.69 ms. Ceci est probablement due au laser qui possède sont propre driver et qui ne désire pas être modulé par un signal. Nous pensons qu'à une certaine tension de monté le laser débute la sortie. La deuxième hypothèse est que le laser prend un certain temps une fois que la bonne tension est à sont entré pour réchauffé et émettre son load.
