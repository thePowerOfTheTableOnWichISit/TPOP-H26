On fait des tests sur l’oscilloscope, le laser est directement envoyé sur le détecteur avec une distance de 0,45 m -> différence de temps de 0,0000000015 s -> 1,5 ns.

La première mesure est 0,638 milliseconde -> 638000 ns. Notre première idée est de trouver le temps que prend le laser pour être détecté à la sortie, puis de trouver le temps à une certaine distance et de soustraire les deux. La mesure à 7,5 cm est de 36800 ns, la grande différence provient sûrement de trucs weird, on va reprendre une mesure plus loin.

On voulait utiliser le sync out du générateur de fonctions pour savoir quand le signal est haut. Cependant, il y a une différence d’environ 10 ns entre la montée du sync et le signal en sortie. On a donc décidé de mesurer la montée du signal directement à l’aide d’un connecteur en T.

Le temps entre la montée du signal envoyé et la détection du laser est d’environ 1,5 ms. Ceci porte à croire qu’un temps non négligeable s’écoule avant que le laser émette. Deux hypothèses sont émises. Premièrement, la tension monte sur une plage de 6,69 ms. Ceci est probablement dû au laser qui possède son propre driver et qui ne désire pas être modulé par un signal. Nous pensons qu’à une certaine tension de montée, le laser débute la sortie. La deuxième hypothèse est que le laser prend un certain temps, une fois que la bonne tension est à son entrée, pour chauffer et émettre.

On a essayé pendant un moment de faire marcher le code Python, mais ça bug à cause d’une connexion API déjà faite.

On fait maintenant un test avec un offset de 6,3 V et une crête de 8 V. Ceci ne fermait jamais le laser. Ceci nous indique qu’une tension de 6,3 V est suffisante pour maintenir l’émission. On a settle sur un offset de 4 V avec un top de 8 V. On remarque que

distance courte 28 cm, distance longue 48 cm

On met le capteur à côté du laser à une distance de 8 cm et on détecte la diffusion à l’aide d’un gain de 70 dB.

Après analyse des données, il paraît bien que Moku:Go offre une fonctionnalité pour faire l’acquisition des données à tous les 40 ns. Ceci est intéressant. On a pris des mesures avec

Avec l’oscilloscope, on prend des mesures tous les 10 centimètres avec un rate de 1 GSa/s.

Les premières données étaient mauvaises parce qu’on avait juste un front. Puis on a essayé d’avoir plus de fronts en augmentant la fréquence du signal à 600 Hz, mais enregistrer ces données fut très gossant car il fallait les enregistrer en bin pour que ce ne soit pas trop gros pour le logiciel. On a ensuite passé une heure à essayer de les transformer en quelque chose de pertinent.

La raison pour laquelle on veut augmenter la fréquence du signal est que présentement, il est très difficile d’avoir une bonne mesure de temps pour la différence entre le sync et la détection. Beaucoup de bruit est présent et il faudrait un large dataset pour en tirer quelque chose de pertinent.

La dernière technique qui fut employée en cette journée ensoleillée est de prendre des données avec un signal de 1 kHz. Ceci permet de sortir une image CSV avec 5 fronts d’onde. Cette donnée n’a pas été utilisée.
